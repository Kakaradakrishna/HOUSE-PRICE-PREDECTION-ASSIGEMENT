# SPLITING INTO INPUT AND OUTPUT COLUMNS

```python
x = df.drop(['House price of unit area'], axis=1)
y = df['House price of unit area']
```

```python
y.head()
```

# SPLITING DATA INTO TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state = 42)
```

```python
print('shape of input-training set:',x_train.shape)
print('shape of output-training set:',y_train.shape)
print('shape of input-testing set:',x_test.shape)
print('shape of output-testing set:',y_test.shape)
```

# TRANSFORMING THE INPUT VARIABLES

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
x_train= scale.fit_transform(x_train)
x_test= scale.transform(x_test)
```

```python
x_train
```

# BUILDING THE REGRESSION MODELS

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

# 1. LINEAR REGRESSION

```python
lm=LinearRegression()
lm.fit(x_train,y_train)


y_pred=lm.predict(x_test)

accuracy_lm=cross_val_score(estimator = lm, X = x_train, y = y_train, cv = 8,verbose = 1)
score_lm=lm.score(x_test,y_test)
r2_lm=r2_score(y_test,y_pred)
mse_lm=mean_squared_error(y_test,y_pred)
```

```python
print(accuracy_lm)
print(score_lm)
print(r2_lm)
print(mse_lm)
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
```

# 2.RANDOMFOREST REGRESSOR

```python
rf = RandomForestRegressor(max_depth=2, random_state=0)
rf.fit(x_train,y_train)


y_pred=rf.predict(x_test)

accuracy_rf=cross_val_score(estimator = rf, X = x_train, y = y_train, cv = 8,verbose = 1)
score_rf=rf.score(x_test,y_test)
r2_rf=r2_score(y_test,y_pred)
mse_rf=mean_squared_error(y_test,y_pred)
```

```python
print(accuracy_rf)
print(score_rf)
print(r2_rf)
print(mse_rf)
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
```

# 3. K-NEAREST-NEIGHBOUR

```python
knn=KNeighborsRegressor(n_neighbors=10)
knn.fit(x_train,y_train)


y_pred=knn.predict(x_test)

accuracy_knn=cross_val_score(estimator = knn, X = x_train, y = y_train, cv = 8,verbose = 1)
score_knn=knn.score(x_test,y_test)
r2_knn=r2_score(y_test,y_pred)
mse_knn=mean_squared_error(y_test,y_pred)
```

```python
print(accuracy_knn)
print(score_knn)
print(r2_knn)
print(mse_knn)
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
```

# 4.DECISIONTREE REGRESSOR

```python
dt = DecisionTreeRegressor(random_state=0)

dt.fit(x_train,y_train)


y_pred=dt.predict(x_test)

accuracy_dt=cross_val_score(estimator = dt, X = x_train, y = y_train, cv = 8,verbose = 1)
score_dt=dt.score(x_test,y_test)
r2_dt=r2_score(y_test,y_pred)
mse_dt=mean_squared_error(y_test,y_pred)
```

```python
print(accuracy_dt)
print(score_dt)
print(r2_dt)
print(mse_dt)
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
```
